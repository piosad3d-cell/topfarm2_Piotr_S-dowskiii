```python
# Ulepszony kod (zgodnie z Krokiem 7)
turbiny = ['T1', 'T2', 'T3', 'T4']
kolory = ['blue', 'red', 'green', 'orange']

plt.figure(figsize=(8, 6))

# Pętla rysująca każdy punkt osobno z innym kolorem i etykietą
for i in range(len(x)):
    plt.scatter(x[i], y[i], color=kolory[i], s=100, label=turbiny[i])
    plt.text(x[i], y[i]+50, turbiny[i], ha='center', fontsize=12)

plt.title("Layout turbin – Zadanie 2.1 z kolorami i etykietami")
plt.xlabel("x [m]")
plt.ylabel("y [m]")
plt.grid(True, linestyle='--', alpha=0.7)
plt.axis("equal")
plt.legend()
plt.show()
```
